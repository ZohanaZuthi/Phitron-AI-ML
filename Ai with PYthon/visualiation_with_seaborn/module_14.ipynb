## Line Plot

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('../media/student_attendance_study_dataset.csv')
df.head()
```

```python
# axes level
sns.lineplot(data=df,x='Study_Hours',y='Attendance_Rate',hue='Gender')
```

```python
# figure level
sns.relplot(data=df,x='Study_Hours',y='Attendance_Rate',hue='Gender',kind='line')
```

```python
# figure level
sns.relplot(data=df,x='Study_Hours',y='Test_Score',hue='Class_Level',kind='line')
```

## Scatter Plot

### To see Correlation 

```python
# axes level
sns.scatterplot(data=df, x='Study_Hours', y="Test_Score", hue='Class_Level', style="Gender")
```

```python
# figure level
sns.relplot(data=df,x='Study_Hours',y='Attendance_Rate',hue='Gender',style='Class_Level',kind='scatter')   
```

## Facet Plot

### figure level multiple plot

```python
# figure level
sns.relplot(data=df,x='Study_Hours',y='Attendance_Rate',col='Gender',kind='scatter')   
```

```python
# figure level
sns.relplot(data=df,x='Study_Hours',y='Attendance_Rate',col='Gender',row='Hostel',kind='scatter')   
```

## Histogram

### Data distribution

```python
# axes level
sns.histplot(data=df,x='Attendance_Rate',hue='Gender')
```

```python
# figure level
sns.displot(data=df,x='Attendance_Rate',col='Gender', kind='hist')
```

## KDE Plot (Kernel Density Estimation)

```python
# figure level
sns.kdeplot(data=df,x='Attendance_Rate', hue='Gender', fill=True)
```

## Count Plot

```python
sns.countplot(data=df,x='Class_Level')
```

```python
sns.countplot(data=df,x='Class_Level',hue='Gender')
```

## BarPlot

### To show aggregaet functions

```python
sns.barplot(data=df, x='Gender', y='Test_Score', hue='Class_Level', errorbar=None)
```

```python
sns.barplot(data=df, x='Gender', y='Test_Score', hue='Class_Level', errorbar=None, estimator=np.max)
```

```python
sns.barplot(data=df, x='Gender', y='Test_Score', hue='Class_Level', errorbar=None, estimator=np.median)
```

## RegPlot(Regression Plot)

### Regression line attached with scatterplot

```python
# axes level
sns.regplot(data=df, x="Study_Hours", y="Test_Score")
```

```python
# figure level
sns.lmplot(data=df, x="Study_Hours", y="Test_Score", hue='Gender')
```

## Pair Plot

### correlation type

```python
pp=df[['Study_Hours','Test_Score','Attendance_Rate','Gender']]
pp
```

```python
sns.pairplot(pp)
```

```python
sns.pairplot(pp, kind='hist')
```

```python
sns.pairplot(pp,hue='Gender')
```

## JointPlot

```python
sns.jointplot(data=pp, x='Study_Hours', y='Test_Score', kind='scatter')
```

```python
sns.jointplot(data=pp, x='Study_Hours', y='Test_Score', kind='kde')
```
