# Class and Objects

```python
# base class/ parent class ( starting with capital letter)

class Phone:
    category="Electronics"
    
    # constructor
    def __init__(self, model,battery,camera,battery_percentage=100):
        self.model=model
        self.battery=battery
        self.camera=camera
        self.battery_percentage=battery_percentage
    
    # methods
    def charge(self,hour):
        print(f"charge completed by {hour } {self.model}")
    
    def capture(self,photo):
        if(self.battery_percentage)<=0:
            print("No charge")
        else:
            self.battery_percentage-=photo
            print(f"photo captured in {self.model}")


class Cooling_Mechanism:
    def __init__(self, cooling_type):
        self.cooling_type=cooling_type
        
    def cooling_on(self):
        print(f"Cooling mechanism {self.cooling_type} is on")
    
    
        
    
```

```python
# creating object/ instance of the class

apple=Phone("iPhone 14 Pro", "Li-ion", "48MP")
blueberry=Phone("Blueberry S1", "Li-ion", "64MP")
motorola=Phone("Motorola G22", "Li-ion", "50MP")
apple.charge(10)
apple.capture(20)
print(apple.battery_percentage)
print(motorola.battery_percentage)

```

# Single Inheritance

```python
# child/derived class
class SmartPhone(Phone):
    def __init__ (self, model, battery,camera, processor):
        super().__init__(model,battery, camera)
        self.processor=processor
    
    def charge(self, hour):
        super().charge(hour)
```

```python
pro=SmartPhone("Samsung S22", "Li-ion", "50MP", "Snapdragon 8 Gen 1")
print(pro.model)
pro.charge(15)
```

# Multiple Inheritance

```python
class Smart_CoolingMode(SmartPhone, Cooling_Mechanism):
    def __init__(self, model, battery,camera, processor, cooling_type):
        SmartPhone.__init__(self, model, battery,camera, processor)
        Cooling_Mechanism.__init__(self, cooling_type)
        
    def charge(self, hour):
        super().charge(hour)
        self.cooling_on()
```

```python
pro_cooling=Smart_CoolingMode("Samsung S22 Ultra", "Li-ion", "50MP", "Snapdragon 8 Gen 1", "Liquid Cooling")
pro_cooling.charge(20)
```

# Polymorphism

```python
class Camera:
    def __init__(self, name):
        self.name=name
        
    # method
    def capture (self):
        print(f"photo captured by Camera")
        
```

```python
class Smart_Capture(Camera):
    def __init__(self, name, resolution):
        super().__init__(name)
        self.resolution=resolution
        
    # method overriding
    def capture(self):
        print(f"photo captured by {self.name} with resolution {self.resolution}")
        
```

```python
class DSLR(Camera):
    def __init__(self, name, resolution):
        super().__init__(name)
        self.resolution=resolution
        
    # method overriding
    def capture(self):
        print(f"photo captured by DSLR")
        
```

```python
camera1=Camera("Generic Camera")
smart_cap=Smart_Capture("Smart Capture", "108MP")
dslr=DSLR("Canon DSLR", "24MP")


camera1.capture()
smart_cap.capture()
dslr.capture()
```

# Encapsulation

```python
# design
class Mobile:
  def __init__(self,name,model,imei):
    self.__name = name
    self.__model = model
    self.__imei = imei # private

  def charge(self):
    print("phone is charging")

  # method imei ta paite pari
  
  def get_imei(self):
    return self.__imei
  
  def name_getter(self):
    return self.__name

  def name_setter(self, new_name):
    self.__name = new_name
```

```python
iphone=Mobile("iPhone", "14 Pro", "123456789012345")
print(iphone.get_imei())
iphone.name_setter("iPhone 14 Pro Max")
print(iphone.name_getter())
```

# Abstraction

```python
from abc import ABC, abstractmethod

class Telephone(ABC):
    @abstractmethod
    def charge(self):
        pass
    

class Sphone(Telephone):
    def charge(self):
        print("Sphone is charging")
        
class Lphone(Telephone):
    def charge(self):
        print("Lphone is charging")
```
