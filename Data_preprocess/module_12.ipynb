## Filtering

```python
import pandas as pd

data = {
    "emp_id": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    
    "Name": [
        "John Smith", "Jane Doe", "Michael Brown", "Emily Davis", "David Wilson",
        "Sarah Miller", "Robert Taylor", "Olivia Anderson", "James Thomas", "Sophia Martinez"
    ],
    
    
    "City": [
        "New York", "New Los Angeles", "Chicago", "Los Houston", "Phoe Los nix",
        "Dallas", "San Diego", "Austin", "New Seattle", "Miami"
    ],
    
    "Department": [
        "HR", "Engineering", "Marketing", "Finance", "Engineering",
        "HR", "Sales", "Engineering", "Management", "Marketing"
    ],
    
    
    "Salary": [50000, 70000, 55000, 60000, 90000, 75000, 48000, 68000, 120000, 52000],
   
}

df=pd.DataFrame(data)
print(df)
```

```python
df.loc[df['City'].str.contains("New")]
```

```python
df.loc[df['City'].str.contains("new")]
```

```python
# case insensitive search
df.loc[df['City'].str.contains("New", case=False)]
```

```python
# Starts with Los
df.loc[df['City'].str.contains(r"^Los", case=False)]
```

```python
# Ends with k
df.loc[df['City'].str.contains(r"rk$", case=False)]
```

```python
# name starts with vowel
df.loc[df['Name'].str.contains(r"^[AEIOU]", case=False)]
```

```python
df.loc[df['City'].str.contains(r"New|Los")]
```

## Adding New Columns


```python
import pandas as pd

df = pd.read_csv("../media/student_grades_with_nulls.csv")

# Create First_Name from full Name
df["First_Name"] = df["Name"].str.split().str[0]

# Create Total from subject columns
df["Total"] = df[["Math", "English", "Science"]].sum(axis=1)

# Save updated dataset as new CSV
df.to_csv("student_grades_updated.csv", index=False)

print(df)
```
