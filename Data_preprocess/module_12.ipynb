## Filtering

```python
import pandas as pd

data = {
    "emp_id": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    
    "Name": [
        "John Smith", "Jane Doe", "Michael Brown", "Emily Davis", "David Wilson",
        "Sarah Miller", "Robert Taylor", "Olivia Anderson", "James Thomas", "Sophia Martinez"
    ],
    
    
    "City": [
        "New York", "New Los Angeles", "Chicago", "Los Houston", "Phoe Los nix",
        "Dallas", "San Diego", "Austin", "New Seattle", "Miami"
    ],
    
    "Department": [
        "HR", "Engineering", "Marketing", "Finance", "Engineering",
        "HR", "Sales", "Engineering", "Management", "Marketing"
    ],
    
    
    "Salary": [50000, 70000, 55000, 60000, 90000, 75000, 48000, 68000, 120000, 52000],
   
}


df=pd.DataFrame(data)
print(df.shape)
print(df)
```

```python
df.loc[df['City'].str.contains("New")]
```

```python
df.loc[df['City'].str.contains("new")]
```

```python
# case insensitive search
df.loc[df['City'].str.contains("New", case=False)]
```

```python
# Starts with Los
df.loc[df['City'].str.contains(r"^Los", case=False)]
```

```python
# Ends with k
df.loc[df['City'].str.contains(r"rk$", case=False)]
```

```python
# name starts with vowel
df.loc[df['Name'].str.contains(r"^[AEIOU]", case=False)]
```

```python
df.loc[df['City'].str.contains(r"New|Los")]
```

## Apply function on DF


### Min_max Scaling

```python
mn=df["English"].min()
mx=df["English"].max()

df["English_Scaled"] = (df["English"] - mn) / (mx - mn)
# or using apply
df["English_Scaled"] = df["English"].apply(lambda x: (x - mn) / (mx - mn) if mx != mn else 0)
```

```python
def grading_system(marks):
    if marks >= 260:
        return "A+"
    elif marks >= 250:
        return "A"
    elif marks >= 240:
        return "B"
    elif marks >= 230:
        return "C"
    else:
        return "F"
    
df["Grade"] = df["Total"].apply(grading_system)
df
```

```python
def marking_syste(df):
    a=df["English"]*2
    b=df["Math"]*2
    c=df["Science"]*2
    
    return a+b+c
df["Exceptional_Total"] = df.apply(marking_syste, axis=1)
df
```

## Group by pandas


```python
import pandas as pd

df = pd.read_csv("../media/thesis_students_marks.csv")

df
```

```python
instructor_group = df.groupby("instructor")

summary = instructor_group[["proposal_mark", "mid_mark", "final_mark"]].agg(["mean", "min", "max"])

summary
```
