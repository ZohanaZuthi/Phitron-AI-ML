## Split Dataset

### array_split for non equal split, split for equal split

```python
a=np.random.randint(1,10,size=(2,3))
print(a)
splitted_array=np.array_split(a,3)
print(splitted_array)
splitted_array=np.split(a,2)
print(splitted_array)
```

## Comparison

```python
x=np.array([1,2,3,4,5])
y=np.array([5,4,3,2,1])

greater=x>y
print(greater)

equal=x==y
print(equal)

print(np.all(greater))
print(np.any(greater))
```

```python
mat=np.array([[4,1,6],[9,2,3],[7,3,9]])
# horizon sort
hor_sort=np.sort(mat,axis=1)
print(hor_sort)
# vertical sort
ver_sort=np.sort(mat,axis=0)
print(ver_sort)
```
