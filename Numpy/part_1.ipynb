## Nd-Array
N dimensional array.


```python
arr=[1,2,3]
print(arr*5)

arr=[x*5 for x in arr]
print(arr)
```

```python
import numpy as np
arr=np.array(arr)
print(arr.dtype)

arr1=np.array([1.2,3.4,6.5])
print(arr1.dtype)
```

```python
# integer into float it is called upcasting
arr=np.array([1,2,3.2])
print(arr.dtype)
# upcasting to string
arr2=np.array([1,2,3,"hello"])
print(arr2.dtype)
```

```python
# let's try with tuple
tup=(1,2,3)
arr=np.array(tup)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.ndim)
```

```python
# let's try with set
st={1,2,3}
arr=np.array(st)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.ndim)

# here set is taken as object , so the data type can not be changed
#  so first change it to list and then to array
st={1,2,3}
arr=np.array(list(st))
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.ndim)
```

```python
# let's try with dictionary
d={1:"one", 2:"two", 3:"three"}
key=d.keys()
values=d.values()
items=d.items()
arr=np.array(items)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.ndim) 
```

## Creating random values

```python
r=np.random.rand(2,3)
# 0 to 1 float value
arr=np.array(r)
print(arr)
print(type(arr))        
print(arr.dtype)
print(arr.ndim)
```

```python
r=np.random.randint(1,10,(2,3))
# 0 to 1 float value
arr=np.array(r)
print(arr)
print(type(arr))        
print(arr.dtype)
print(arr.ndim)
```

```python
# to generate floating value

r=np.random.uniform(1,10,(2,3))
# 0 to 1 float value
arr=np.array(r)
print(arr)
print(type(arr))        
print(arr.dtype)
print(arr.ndim)
```

## array creating with range

```python
# np.arrange(first,end, step) 
arr = np.arange(1,10,2)
print(arr)
print(type(arr))
print(arr.dtype)
```

```python
arr=np.arange(1,10,1).reshape(3,3)
print(arr)
print(type(arr))
print(arr.ndim)
```

```python
# np.linespace(start,stop(included),koita)

arr = np.linspace(1,10,5)
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.ndim)
```

```python
# np.logspace(start,stop(included),koita)

arr=np.logspace(0,4,5)

print(arr)
print(type(arr))
print(arr.dtype)
print(arr.ndim)
```

```python
# np.logspace(start,stop(included),koita)

arr=np.logspace(0,4,6,base=2)

print(arr)
print(type(arr))
print(arr.dtype)
print(arr.ndim)
```

## getting a column or row

```python
arr=[1,2,3,4,5,6,7,8,9]
arr=np.array(arr).reshape(3,3)
print(arr)
print(type(arr))
print(arr.dtype)
```

```python
col_0=arr[:,0:1]
print(col_0)
```

```python
row_0=arr[0:1,:]
print(row_0)
```

```python
portion=arr[::,0:2]
print(portion)
```

```python
arr=[1,2,3,4,5,6,7,8,9]
arr=np.array(arr)
arr1=arr[0:5]
# here in this way the array is not getting copied just indexing is happening
#  so if any changes brought in arr1 it will also reflect in arr
print(arr1)
arr1[0]=100
print(arr1)
print(arr)
```

```python
arr_copy=arr[1:5].copy()
print(arr_copy)
arr_copy[0]=200
print(arr_copy)
print(arr1)
```

## Advanced Indexing

```python
lst=np.array([1,2,3,4,5])
values=lst[[2,4]]
print(values)
```

```python
arr=np.array([[1,2,3],[4,5,6]])
print(arr[[0,1],[1,2]])
print(arr[arr<2])

arr[arr<3]=0
print(arr)
```
