## 2D Line Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hours=[2,3,4,1,7,9,4]
days=[1,2,3,4,5,6,7]
plt.plot(days,hours)
```

```python
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.title('Study hours of a student in a week')

plt.plot(days,hours, marker='o', linestyle='--', color='r')
```

```python
hours_1=[2,3,4,1,7,9,4]
hours_2=[1,2,3,4,5,6,7]
days=[1,2,3,4,5,6,7]

plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.title('Study hours of a student in a week')

plt.plot(days,hours_1,label='Student 1')
plt.plot(days,hours_2,label='Student 2' )

plt.legend('lower left')
```

## 2D line plot from pandas data

```python
df=pd.read_csv("../media/student_enrollment_subjects.csv")
df
```

```python
df["Enrollment_Date"] = pd.to_datetime(df["Enrollment_Date"])
```

```python
plt.plot(df["Enrollment_Date"].dt.month)
plt.xlabel("Row Number")
plt.ylabel("Enrollment Month")
plt.show()
```

```python
cse_df = df[df["Department"] == "CSE"]
print(cse_df)
```

```python
cse_yearly_trend = cse_df.groupby("Enrollment_Year").size()

print(cse_yearly_trend)
```

```python
plt.plot(cse_yearly_trend.index, cse_yearly_trend.values, marker="o")

plt.xlabel("Year")
plt.ylabel("Number of CSE Enrollments")
plt.title("Trend of Choosing CSE Subjects by Year")
```

## Line Plor(color, line style and marker)

```python
df=pd.read_csv("../media/yearly_category_enrollment.csv")
df
```

```python
plt.plot(df['Year'],df['Programming'], marker='o', label='Programming', color='r',linewidth=5,linestyle='--')
plt.plot(df['Year'],df['Data Science'], marker='s', label='Data Science', color='g')
plt.plot(df['Year'],df['BBA'], marker='^', label='BBA', color='b')
plt.xlabel('Year')
plt.ylabel('Enrollment Count')
plt.title('Yearly Enrollment Trends by Category')
plt.legend()
plt.show()
```
